# HUMAN FACE DETECTOR




```python
# importing the required libraries
import cv2
from google.colab.patches import cv2_imshow
```

```python
# load and view image
face_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
img = cv2.imread('/content/human pic.webp')
cv2_imshow(img)
```

```python
# Converting the image to gray scale
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
cv2_imshow(gray)
```

```python
faces = face_cascade.detectMultiScale(gray)   # Returns the co-ordinates, width and height of rectange
print(faces)
```

```python
# Draw and display the rectangle over face
for (x,y,w,h) in faces:
  cv2.rectangle(img, (x,y), (x+w,y+h), (0,100,150), 2)

cv2_imshow(img)
```

# HUMAN EYES DETECTOR

```python
# Loading the classifier
eyes_cascade = cv2.CascadeClassifier('/content/haarcascade_eye.xml')

# applying the classifier and getting the co-ordinates
eyes = eyes_cascade.detectMultiScale(gray)
print(eyes)

# Drawing rectangles on the eyes based on co-ordinates
for (x,y,w,h) in eyes:
  cv2.rectangle(img, (x,y), (x+w,y+h), (100,150,100), 2)

# Displaying the output image
cv2_imshow(img)
```

# CREATING FUNCTION TO DETECT FACE

```python
def fun_to_detect(img_path):
  # Loading the classifier
  face_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')

  # Loading the image
  img = cv2.imread(img_path)

  # Converting the image to gray scale
  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

  # applying the classifier and getting the co-ordinates
  faces = face_cascade.detectMultiScale(gray)

  # Drawing rectangles on the eyes based on co-ordinates
  for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w,y+h), (90,150,200), 2)
  
  # Displaying the output image
  cv2_imshow(img)
```

```python
# Calling the Function
fun_to_detect('/content/5d020eb457ae2.jpg')
```
